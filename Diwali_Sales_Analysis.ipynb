# Exploratory Data Analysis


```python
df.columns
```

```python
b=sns.countplot(x="Gender",data=df)
```

```python
# no of purchases by male and female
b=sns.countplot(x="Gender",data=df)
for bars in b.containers:
    b.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
    

```

# Age

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

# from above graph we can see that most of the buyers are of age between 26-35

# state

```python
df.columns
```

```python
# total number of orders from op 10 states
state_sales=df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=state_sales,x='State',y='Orders')
```

```python
state_amount=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=state_amount,x='State',y='Amount')
```

#  from above graphs we can see that most of the orders and totalsales/amount
# is from uttar pradesh,maharastra and karnataka respectively

# marital status

```python
ax=sns.countplot(x='Marital_Status',hue='Gender',data=df)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_status_sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=marital_status_sales,x='Marital_Status',y='Amount',hue="Gender")
```

# from above graphs we can see that most of the buyers are married(women) and they have high purchasing power

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Occupation_sales=df.groupby('Occupation',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=Occupation_sales,x='Occupation',y='Amount')
```

# from above graphs we can see that most of the workers are working in It,healthcare and Aviation

# Product category

```python
sns.set(rc={'figure.figsize':(25,7)})
ax=sns.countplot(x='Product_Category',data=df)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Product_sales=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=Product_sales,x='Product_Category',y='Orders')
```

```python
Product_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=Product_sales,x='Product_Category',y='Amount')
```

# From above we can see that most of the products sold are from food,clothing and electronics

```python
sns.set({'figure.figsize':(20,5)})
product_id_sales=df.groupby(["Product_ID"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=product_id_sales,x='Product_ID',y='Orders')
```
